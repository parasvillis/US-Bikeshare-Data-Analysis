```python
import pandas as pd
from datetime import *
import numpy as np
import time
from statistics import mode
```

```python
def load_data():
    city = ''
    while city.lower() not in ['chicago', 'new york', 'washington']:
        city = input("Enter a city for which you want to get data from: ('chicago', 'washington' or 'new york')")
        if city.lower() == 'chicago':
            return 'chicago.csv'
        elif city.lower() == 'new york':
            return 'new_york_city.csv'
        elif city.lower() == 'washington':
            return 'washington.csv'
        else:
            print('Invalid City')

def load_period():
    time_period = ''
    while time_period.lower() not in ['month', 'day', 'none']:
        time_period = input('\nHow do you want to filter data (by "month", "day" or "none")')
        if time_period.lower() not in ['month', 'day', 'none']:
            print('Invalid month.')
    return time_period

def filter_month():
    month_input = ''
    months_dict = {'january': 1, 'february': 2, 'march': 3, 'april': 4,
                   'may': 5, 'june': 6}
    while month_input.lower() not in months_dict.keys():
        month_input = input('\nWhich month? January, February, March, April,'
                            ' May, or June?\n')
        if month_input.lower() not in months_dict.keys():
            print('Invalid month')
    month = months_dict[month_input.lower()]
    return ('2017-{}'.format(month), '2017-{}'.format(month + 1))

def filter_day():
    this_month = filter_month()[0]
    month = int(this_month[5:])
    valid_date = True
    while valid_date:    
        day = int(input('\nEnter a day of your choice.. (Please enter an integer).\n'))
        try:
            start_date = datetime(2017, month, day)
            valid_date = False
        except Exception as e:
            print(e)
    end_date = start_date + timedelta(days=1)
    return (str(start_date), str(end_date))

def countX(lst, x):
    count = 0
    for ele in lst:
        if (ele == x):
            count +=  1
    return count

def popular_month(df):
    start_time = time.time()
    
    months = ['January', 'February', 'March', 'April', 'May', 'June']
    index = int(df['Start Time'].dt.month.mode())
    most_pop_month = months[index - 1]
    print('The most popular month is {}.'.format(most_pop_month))
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)

def popular_day(df):
    start_time = time.time()
    
    days_of_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday',
                    'Saturday', 'Sunday']
    index = int(df['Start Time'].dt.dayofweek.mode())
    most_pop_day = days_of_week[index]
    print('The most popular day of week for start time is {}.'.format(most_pop_day))
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)

def popular_hour(df):
    start_time = time.time()
    
    most_pop_hour = int(df['Start Time'].dt.hour.mode())
    if most_pop_hour == 0:
        am_pm = 'am'
        pop_hour_readable = 12
    elif 1 <= most_pop_hour < 13:
        am_pm = 'am'
        pop_hour_readable = most_pop_hour
    elif 13 <= most_pop_hour < 24:
        am_pm = 'pm'
        pop_hour_readable = most_pop_hour - 12
    print('The most popular hour of day for start time is {}{}.'.format(pop_hour_readable, am_pm))
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
    
def convert_time(sec):
        days = sec // (24 * 3600)
        sec %= (24 * 3600)
        hour = sec // 3600
        sec %= 3600
        minutes = sec // 60
        sec %= 60
        return ([days, hour, minutes, sec])


def trip_duration(df):
    print('\nCalculating Trip Duration...\n')
    start_time = time.time()

    # display total travel time
    sec = df['Trip Duration'].sum()
    res = convert_time(sec)
    print("\n\nTotal time of ride: {} days {} hours {} minutes {} seconds".format(res[0], res[1], res[2], res[3]))

    # display mean travel time
    mean = np.mean(df['Trip Duration'])
    print("Average time {}".format(mean))

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)

def popular_stations(df):
    print('\nCalculating The Most Popular Stations and Trip...\n')
    start_time = time.time()

    # display most commonly used start station
    st_lst = []
    for i in df['Start Station']:
        st_lst.append(i)
    print("\n\nThe most common Start station is: {} and has visited {} times".format(mode(st_lst), countX(st_lst, mode(st_lst))))

    # display most commonly used end station
    en_lst = []
    for j in df['End Station']:
        en_lst.append(j)
    print("The most common End station is: {} and has visited {} times".format(mode(en_lst), countX(en_lst, mode(en_lst))))

    
    # display most frequent combination of start station and end station trip
    try:
        trp_lst = []
        for i, j in zip(df['Start Station'], df['End Station']):
            trp_lst.append(i+j)
        print("\n\nMost Popular Trip is {} and is taken {} times".format(mode(trp_lst), countX(trp_lst, mode(trp_lst))))
    except Exception:
        print("No most popular Trip")
        

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)

def user_type(df):
    print('\nCalculating User Stats...\n')
    start_time = time.time()

    # Display counts of user types
    sub, cus, nan1 = 0,0,0
    for u in df['User Type']:
        if u == 'Subscriber':
            sub += 1
        elif u == 'Customer':
            cus += 1
        else:
            nan1 += 1
    print("\nTotal number of Subscribers: {}".format(sub))
    print("Total number of Customers: {}".format(cus))
    print("{} people has a guest account yet.".format(nan1))
    

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
    
def gender(df):
    print('\nCalculating User Stats...\n')
    start_time = time.time()

    male, female, nan = 0,0,0
    for g in df['Gender']:
        #print(g)
        if g == 'Male':
            male += 1
        elif g == 'Female':
            female += 1
        else:
            nan += 1
    print("There are total {} males.".format(male))
    print("There are total {} females.".format(female))
    print("{} people has not specified their gender yet.".format(nan))

    Sm = Sf = Cm = Cf = 0
    for i, j in zip(df['User Type'], df['Gender']):
        #print(i, j)
        if i == 'Subscriber' and j == 'Male':
            Sm += 1
        elif i == 'Subscriber' and j == 'Female':
            Sf += 1
        elif i == 'Customer' and j == 'Male':
            Cm += 1
        elif i == 'Customer' and j == 'Female':
            Cf += 1
    print("Male Subscribers: {}".format(Sm))
    print("Female Subscribers: {}".format(Sf))
    print("Male Customers: {}".format(Cm))
    print("Female Customers: {}".format(Cf))

    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)

def birth_years(df):
    start_time = time.time()
    
    earliest = int(df['Birth Year'].min())
    latest = int(df['Birth Year'].max())
    mode = int(df['Birth Year'].mode())
    print("Earliest birth year is {}".format(earliest))
    print("Latest birth year is {}".format(latest))
    print("Most popular birth year is {}".format(mode))
    
    print("\nThis took %s seconds." % (time.time() - start_time))
    print('-'*40)
    
    
    
    
    
    
def display_data(df):
    
    def is_valid(display):
        if display.lower() == 'yes' or 'no' :
            return True
        else:
            return False
    start = 0
    end = 5
    counter = True
    while counter:
        display = input("\nWould you like to view individual trip data? Enter 'yes' or 'no'.")
        counter = is_valid(display)
        if counter:
            break
        else:
            print("Invalid input")
    if display.lower() == 'yes':
        print(df[df.columns[0:]].iloc[start:end])
        display_more = ''
        while display_more.lower() != 'no':
            counter2 = True
            while counter2:
                display_more = input("\nDo you wanna see few more? Enter 'yes' or 'no'.")
                counter2 = is_valid(display_more)
                if counter2:
                    break
                else:
                    print("Invalid input")
            if display_more.lower() == 'yes':
                start += 5
                end += 5
                print(df[df.columns[0:]].iloc[start:end])
            elif display_more.lower() == 'no':
                break

    


def main():
    city = load_data()
    
    df = pd.read_csv(city, parse_dates = ['Start Time', 'End Time'])

    time_period = load_period()
    if time_period == 'none':
        df_filtered = df
    elif time_period == 'month' or time_period == 'day':
        if time_period == 'month':
            filter_lower, filter_upper = filter_month()
        elif time_period == 'day':
            filter_lower, filter_upper = filter_day()
        df_filtered = df[(df['Start Time'] >= filter_lower) & (df['Start Time'] < filter_upper)]

    if time_period == 'none':
        popular_month(df_filtered)
        
    if time_period == 'none' or time_period == 'month':
        popular_day(df_filtered)

    popular_hour(df_filtered)
    
    trip_duration(df_filtered)

    popular_stations(df_filtered)
        
    user_type(df_filtered)
    
    if city == 'chicago.csv' or city == 'new_york_city.csv':
        gender(df_filtered)
        
        birth_years(df_filtered)
    
    display_data(df_filtered)
    
    restart = input("\nWould you like to restart? Enter 'yes' or 'no'.\n")
    while restart.lower() not in ['yes', 'no']:
        print("Invalid input. Please type 'yes' or 'no'.")
        restart = input("\nWould you like to restart? Enter 'yes' or 'no'.\n")
    if restart.lower() == 'yes':
        main()


if __name__ == "__main__":
    main()
```
